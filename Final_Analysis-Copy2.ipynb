```python

def getDistanceTo_reasonable(row,lat0,lon0) :
    deglen = 110.25
    x = row["lat"] - lat0
    y = (row["lon"] - lon0) * math.cos(lat0)
    return deglen * math.sqrt(x*x + y*y)
```

```python

alc_fight_df = reasonable_df.loc[reasonable_df["crime_type"].isin(["*fight / battery*", "*alcohol*"])]
len(alc_fight_df)
```

```python
binNum = 25
binSize = int(len(alc_fight_df)/binNum)

for i in range(14,binNum) :
    n=0
    print(i)
    for row in alc_fight_df.itertuples() :
        n=n+1
        if(n<i*binSize):
            continue 
        if(n>=(i+1)*binSize) :
            break
        try :
            business_reasonable_df[row[0]] = business_reasonable_df.apply(getDistanceTo_reasonable,axis=1,args=(row[3],row[2],))
        except :
            print(row[0], "Did not work")
    business_df_csv = business_reasonable_df.to_csv (r'business_alc_fight_df.csv', header=True) #Don't forget to add '.csv' at the end of the path
```
